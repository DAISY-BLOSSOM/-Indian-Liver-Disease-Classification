```python
X = df.drop(['Dataset','Gender'],axis=1)
y = df.Dataset.values
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2,random_state=15)

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

```

#### Applying ML Algo. Before Balance data

#### Decision Tree

```python
dt=DecisionTreeClassifier(max_depth=4,random_state=42)
dt.fit(X_train,y_train)
y_pred=dt.predict(X_test)
acc=accuracy_score(y_test,y_pred)
print("Accuracy obtained by Decision Tree model: {:.2f} %".format(acc*100))
dt1=acc
```

```python
from sklearn.metrics import confusion_matrix

sns.heatmap(confusion_matrix(y_test,  y_pred),annot=True,fmt="d")
```

#### KNN

```python
from sklearn.neighbors import KNeighborsClassifier

k= range(1,100,2)
testing_accuracy=[]
training_accuracy=[]
score = 0

for i in k:
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    y_predict_train = knn.predict(X_train)
    training_accuracy.append(accuracy_score(y_train,y_predict_train))
    y_predict_test =knn.predict(X_test)
    acc_score = accuracy_score(y_test,y_predict_test)
    testing_accuracy.append(acc_score)
    if score<acc_score:
        score=acc_score
        best_k=i
        yy1 = y_test
        yy2 = y_predict_test
```

```python
sns.lineplot(k,training_accuracy)
sns.scatterplot(k,training_accuracy)
sns.lineplot(k,testing_accuracy)
sns.scatterplot(k,testing_accuracy)
plt.legend(['training_accuracy','testing_accuracy'])
print("The Best K value is",best_k)
print("Accuracy obtained by KNN model: ",score*100 )
knn1=score
```

```python
sns.heatmap(confusion_matrix(yy1, yy2),annot=True,fmt="d")
```

#### Perceptron

```python
mlp = MLPClassifier(
    max_iter=1000,
    alpha=0.1,
    activation='logistic',
    solver='adam',
    random_state=15
)
mlp.fit(X_train, y_train)

mlp_predict = mlp.predict(X_test)
print('Accuracy obtained by MLP: {:.2f}%'.format(accuracy_score(y_test, mlp_predict) * 100))
pr1=accuracy_score(y_test, mlp_predict)
```

```python
sns.heatmap(confusion_matrix(y_test, mlp_predict),annot=True,fmt="d")
```

#### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
LogisticRegressionScore = lr.score(X_test, y_test)
print("Accuracy obtained by Logistic Regression model:",LogisticRegressionScore*100)
lr1=LogisticRegressionScore
```

```python
y_pred_lr = lr.predict(X_test)
sns.heatmap(confusion_matrix(y_test, y_pred_lr ),annot=True,fmt="d")
```

#### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators = 100, max_depth=9, random_state=42) 
                        # in n=5 ,acc= 79.3103448275862  ==> but in doc the new range is from 10 to 100
rfc.fit(X_train,y_train)
y_pred=rfc.predict(X_test)
RandomForestClassifierScore = metrics.accuracy_score(y_test, y_pred)
print("Accuracy obtained by Random Forest Classifier model:",RandomForestClassifierScore*100)
rnf1=RandomForestClassifierScore
```

```python
y_pred_rfc = rfc.predict(X_test)
sns.heatmap(confusion_matrix(y_test, y_pred_rfc ),annot=True,fmt="d")
```

#### Support Vector Machine

```python
from sklearn.svm import SVC
svm = SVC(kernel= "linear",C=0.025, gamma=0.01)
svm.fit(X_train,y_train)
y_pred_svm = svm.predict(X_test)
ss = svm.score(X_test, y_test)
print("Accuracy obtained by Support vector Machine model:",ss*100)
svm1=ss
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred_svm ),annot=True,fmt="d")
```
